### Квантили и их частные случаи

<br>Квантиль
<br>Число X является α-квантилем набора данных, если оно делит этот набор данных
таким образом, что α% наблюдений меньше или равны  X и (100−α)%
наблюдений больше или равны X.
<br>Перцентиль — это то же самое, но в процентах

Алгоритм расчета квантиля
- Отсортируйте набор данных по возрастанию.
- Найдите позицию квантиля по формуле: n⋅α, где n — количество элементов в наборе,α — доля, которая нас интересует.
- Определите значение квантиля:
- Если позиция квантиля — целое число, α-квантиль равен значению, которое соответствует этой позиции в упорядоченном наборе данных.
- Если позиция квантиля — дробное число, возьмите среднее значение между двумя ближайшими соседями.


<br>Квартиль - это частный случай квантилей. 
<br>Q1 (первый квартиль) — это 0.25 квантиль
<br>Q2 (второй квартиль) — это 0.5 (медиана)
<br>Q3 (третий квартиль) — это 0.75 квантиль.

<br>Межквартильный размах IQR это ядро данных, которое лежит между Q1 и Q3.
<br>Выброс 1.5 * IQR 
<br>Найти выброс:
- Отсортировать данные в возрастающем порядке.
- Вычислить первый квартиль Q1 (0.25-квантиль) и третий квартиль Q3 (0.75-квантиль).
- Определить границы выбросов: нижняя граница: Q1−1.5 * IQR, верхняя граница: Q3 + 1.5 * IQR
- Отсеять все значения, которые лежат за пределами этих границ, — они и считаются выбросами.

### Дисперсия, формула, опредление

Дисперсия — это статистический показатель, который описывает разброс значений в наборе данных относительно среднего значения.

$$
Var(X) = \frac{1}{n-1} \sum_{i=1}^{n}(x_i - \overline{x})^2
$$
Стандартное отклонение - корень из дисперсии
$$
std = \sqrt{Var(X)}
$$

## Ковариация и корреляция

### Ковариация

Ковариация — это мера совместной изменчивости двух величин.<br>
Ковариация показывает только прямое соотношение, то есть линейную зависимость, когда одна величина меняется пропорционально изменению другой. <br>
Коэффициент ковариации - помогает количественно измерить взаимосвязь между двумя величинами.
$$
Cov(X, Y) = \frac{1}{n-1} \sum_{i=1}^{n} (x_i - \overline{x}) (y_i - \overline{y})
$$ 

### Коэффициент корреляции Пирсона

$$
r_XY = \frac{Cov(X,Y)}{std_X * std_Y}
$$
Важное:
- Коэффициент корреляции Пирсона не зависит от масштаба величин.
- Значение этого коэффициента всегда лежит в универсальном диапазоне от -1 до 1.
- Если значение положительно (находится от 0 до 1), то связь между величинами положительная.
- Если значение отрицательно (находится между −1 до 0), то связь между величинами отрицательная.
- Чем больше абсолютное значение коэффициента корреляции, тем сильнее связь.
- Если значение коэффициента корреляции Пирсона около, то линейной связи между величинами нет или она очень слабая.
- Коэффициент корреляции Пирсона — мера линейной зависимости между двумя величинами. Это значит, что он адекватно оценивает силу и направление взаимосвязи только в случаях, когда связь между величинами может быть отражена прямой линией.
- При наличии выбросов использовать коэффициент корреляции Пирсона нельзя!
- Надо либо удалять выбросы из данных, либо использовать другие методы, которые более устойчивы к выбросам. Например, ранговый коэффициент корреляции Спирмана и Кендала (также они могут применяться при наличии нелинейной зависимости).

## Случайная величина, вероятность, ф-ция вероятности

[CONTENT](#CONTENT)<br>

Случайная величина — это переменная, значение которой можно узнать только в результате проведения измерения.<br>
Вероятность события — это отношение числа случаев, когда событие произошло, к общему числу испытаний или наблюдений.<br>
Функция вероятности определяет вероятность того, что случайная величина примет определённое значение. P(X=x)<br>
Эмпирическая функция распределения определяет вероятность того, что случайная величина примет значение, меньшее или равное заданному.  <br> 
$$
\hat{F}(x) = P(X \leq{x})
$$
Например, если считаем в отношении
$$
x_3
$$
$$
\hat{F}(x_3) = P(X=x_1) + P(X=x_2) + P(X=x_3)
$$

## Равномерное дискретное распределение

[CONTENT](#CONTENT)<br>

Равномерное дискретное распределение — это тип теоретического распределения, в котором каждое возможное значение случайной величины X имеет одинаковую вероятность и лежит в пределах от a до b.
<br><br>Равномерное распределение иногда используют как первый подход. Когда не хватает информации, логично предположить вероятности всех исходов одинаковыми. 
<br><br>Равномерное распределение определяется всего двумя параметрами: минимальным и максимальным значениями. 

$$
X∼U(a, b) 
$$

### Функция вероятности равномерного распределния

<b>Функция вероятности равномерного распределния</b>
$$
P(X=x_1) = P(X=x_2) = ... = P(X=x_n)
$$
$$
P(X=x_1) + P(X=x_2) + ... + P(X=x_n) = 1
$$
$$
n⋅P(X=x)=1
$$
$$
P(X=x) = \frac{1}{n}
$$
где n = b-a +1 - количество возможных значений X$$$$

### Функция распределения равномерного распределения

<b>Функция распределения</b> определяет вероятность того, что случайная величина примет значение меньше или равное заданному.
$$
F(x) = P(X \leq{x}) 
$$
Здесь нет значка галочки над F, так как это теоретическое, а не эмпирически рассчитанное распредение.
$$
F(x)=(x−a+1)⋅P(X=x)= \frac{x-a+1}{n}
$$

### Математическое ожидание равномерного распределения

Математическое ожидание обозначает среднее значение случайной величины.

Математическое ожидание — это взвешенное среднее значение случайной величины, где веса представляют собой вероятности возможных значений этой случайной величины.

Общая формула
$$
E(X) = \sum_{i}^{n} x_i * P(x=x_i)
$$

Формула для равномерного распределения
$$
E(X) = \frac{a+b}{2}
$$

### Дисперсия равномерного распределения

$$
Var(X) = \frac{n^2 - 1}{12}
$$

<b>Непрерывная случайная величина</b> — это тип случайной величины, которая может принимать любое значение внутри определённого интервала.

### Функция плотности вероятности 
— это функция, которая описывает вероятность того, что непрерывная случайная величина примет значение в определённом интервале. Она обозначается как

$$


### Нормальное распределение
— это тип теоретического распределения, в котором значения в основном сосредоточены вокруг среднего. Это распределение имеет форму колокола и описывается двумя параметрами: средним значением μ  и дисперсией σ^2.

### Стандартное нормальное распределение
— частный случай нормального распределения, когда μ=0, σ=1.

### Распределение выборочных средних
— это распределение, показывающее, какие значения принимает среднее значение случайной выборки из генеральной совокупности при многократном повторении эксперимента.

## Статистическая гипотеза

[CONTENT](#CONTENT)<br>

В контексте статистического анализа гипотеза — это предположение, которое мы делаем о параметрах нашего распределения.

Нулевая гипотеза, обозначаемая как H0, это исходное предположение о данных.
Альтернативная гипотеза, обозначаемая как H1, представляет собой предположение, которое противоположно нулевой гипотезе.

Критические значения формируют область допустимых значений для статистики, при условии, что нулевая гипотеза верна (например, 1.96 для 95% z-распределения).

Ошибка первого рода — это ситуация, когда исследователь отвергает нулевую гипотезу, хотя она на самом деле верна.

Уровень значимости α — это вероятность совершения ошибки первого рода, то есть отклонить нулевую гипотезу, когда она верна.

Чем ниже уровень значимости, тем более консервативен тест, то есть он более склонен к поддержанию нулевой г

Уровень значимости:

0.001 = левле и правое критическое значения -/+ 2.58

0.05  = -/+ 1.96

0.1 = -/+ 1.645

Порядок действий в рамках Z-теста:
- Сформулировать H0 и Н1
- Выбрать α, найти соответствующие критические значения.
- Вычислить выборочное среднее (sem = std / sqrt(n)).
- Вычислить z-статистику (z = (x_vibor - mean_gen_sov) / sem).
- Сравнить z-статистику с критическими значениями.

Принять решение: 
Если z-статистика выходит за пределы критических значений, отклонить нулевую гипотезу.
Если z-статистика не выходит за пределы критических значений, принимаем решение, что отклонить нулевую гипотезу нельзя.

<b> Метод обратной стандартизации</b> 

Используется для расчета критических значений при проведении эксперимента, например при 0.05 Альфа

Левая граница: 
μ - 1.96 * SEM

Правая граница:
μ + 1.96 * SEM

## 9

## P-value

[CONTENT](#CONTENT)<br>

<b>P-value</b> — вероятность получить определённое или ещё более экстремальное значение статистического критерия при условии, что нулевая гипотеза верна.

Даже если нулевая гипотеза на самом деле верна, всегда есть некоторый шанс, что к нам в выборку случайно попадут необычные наблюдения и это повлияет на полученное значение статистического критерия. И p-value — это вероятность того, насколько вероятно обнаружить текущие или более экстремальные значения статистики, если нулевая гипотеза верна. 

<b>Порядок расчета P-value
- Ставим нулевую гипотезу, где в том числе делаем предподложение о среднем генеральной совокупности
- Считаем стандартную ошибку среднего
- Считаем z значение по полученной выборке (среднее по выборке минус предполагаемое значение среднего совокупности и делим на SEM)
- По z зачению считаем p-value
- По желанию считаем с помощью обратной стандартизации границы допустимых щначений в пределах нулевой гипотезы

### Ошибка первого и второго рода

Ошибка первого рода, 
α — это ситуация, когда исследователь отвергает нулевую гипотезу, хотя она на самом деле верна.

Ошибка второго рода, β — это ситуация, когда исследователь не отвергает нулевую гипотезу, хотя на самом деле верна альтернативная.

Мощность статистического теста, 1−β — это вероятность правильно отклонить нулевую гипотезу, когда она действительно неверна.

Иными словами, это вероятность обнаружить эффект, если он действительно существует.

Мощность теста зависит от нескольких факторов, включая размер выборки и величину эффекта. Большие выборки и большие эффекты обычно ведут к большей мощности. Мощность теста следует рассматривать ещё на этапе планирования исследования, так как она помогает определить необходимый размер выборки.

## T-test и распределение Стьюдента

[CONTENT](#CONTENT)<br>

Оценка стандартной ошибки (англ. Estimated Standard Error, ESE) 

— это стандартное отклонение выборочных средних, вычисленное на основе стандартного отклонения выборки.

Отличается от ошибки среднего тем, что рассчитывается на основании дисперсии выборки, а не дисперсии ген. совокупности, которая в большинстве случаев нам неизвестна. 

$$
ESE = \frac{S_x}{\sqrt{n}}
$$

Теоретические распределения, построенные с помощью SE и ESE, похожи и чем больше размер выборки, тем больше они похожи.

Использование ESE вносит дополнительную неопределённость в нашу оценку погрешности. Ведь теперь мы используем для расчёта погрешности величину, которая сама является оценкой и имеет свою погрешность.

$$
t = \frac{\overline{x} - \mu}{\sqrt{n}}
$$

t - статистика формирует распредление Стьюдента которое зависит от степеней свобод и имеет более жирные хвосты, особенно при распределениях, где n < 30

df - degree of freedom, сетпени свободы равные  n-1

Степени свободы формируют распредление Стьюдента и влияют на его форму и разброс концов. При малых выборках доверительный интервал такого распределения шире и критические значения борлее удаленные от центра распредения.

$$
H_1 : \mu_1 - \mu_2 \neq 10
$$
